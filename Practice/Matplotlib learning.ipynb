```python
import matplotlib.pyplot as plt
```

# Bar Charts

```python
x = ["Python" , "C", "C++", "Java"]
y = [80, 60, 70, 74] # popularity values
z = [25, 90, 45, 35] # testing data 1
```

```python
plt.xlabel("Languages", fontsize=20)
# Fontsize - x lebel's font size 

plt.ylabel("Popularity %", fontsize=15)
# Fontsize - y label's font size

plt.title("Programming Languages Popularity", fontsize=30, pad=30)
# Fontsize - Chart's main title font size

# Bar chart for popularity
plt.bar(x, y, width=0.2, color=["green", "red", "yellow"], align="edge", edgecolor=["black", "white"], linewidth=1, linestyle=":", alpha=0.5, label="popularity")

# Bar chart for testing data 1
plt.bar(x, z, width=1, color="pink", align="center", edgecolor=["black", "white"], linewidth=1, linestyle=":", alpha=0.5, label="testing data 1")
# Width - width of a bar (default - 1)
# Color - one color or list of colors
# Align - ['edge', 'center'] two values for align attribute (Default - center)
# Edgecolor - Bar's border color (also pass list of colors)
# Linewidth - border's width of bar
# Linestyle - Style of bar's border
# Alpha - Make transparent (bars) (Default - 1) (Values : 0 to 1)
# Label - set Label name for particular bar chart (requires to add plt.legend() command)

plt.legend(loc="upper right")
# This commands show labels for different charts (multiple charts)
# Loc - it is for location and one of the values : 
# ['best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'] 

plt.xticks(rotation=45)
# Xticks - This is used to format X-axes ticks
# Rotation - sets x-labels to some angel in degree

plt.grid()
# this method shows grids in chart

plt.show()
```

# Scatter Charts

```python
x = [1,2,3,4,5,6,7]
y = [5,10,7,3,1,0,7]
z = [25,35,70,80,45,56,78]
```

```python
colors = [10, 30, 45, 78, 65, 3, 94] # Values from 0 to 100 for cmap

plt.scatter(x, y, sizes=[500, 300], alpha=0.7, c=colors, marker="*", edgecolor="b", linewidths=0.45, cmap="magma_r")
plt.scatter(x, z, sizes=[500, 300], alpha=0.7, c=colors, edgecolor="b", linewidths=0.45, cmap="inferno")
# Sizes - list of size values for scatter 
# Marker - Change the style of scatter points ['*', '1', '2', '3', '4'] (Default - O)
# cmap - Automatic change the color according to values given in c parameter. cmap's value is a set of color name.
# Reguler values for cmap is 'viridis', 'viridis_r'(for showing reverse color pattern), 'plasma', 'plasma_r', 'inferno', 'magma', 'cividis', 
# showing reverse color list => viridis_r, plasma_r etc.

color_bar = plt.colorbar()
# this method creates a bar which is showing 0 to 100 values with associate color
color_bar.set_label("testing values", fontsize=20)

plt.title("Simple Scatter Chart", fontsize=25)
plt.show()
```
